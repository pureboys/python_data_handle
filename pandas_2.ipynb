```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

df = DataFrame(data=np.random.randint(0,100, size=(8,6)))
df
```

```python
df[0] #df[列索引]
df[[1,2,3]] # 批量多列
df.iloc[0] # df.iloc[行索引]
df.iloc[[1,2]] # 批量多行
df.iloc[1,2] # 取制定元素
df.iloc[[1,2],3] # 批量取出多个元素
```

```python
# 切片
df[0:3] #df[行切片]
df.iloc[:, 0:3] #df.iloc[行切片, 列切片]
```

```python
df.iloc[2,4] = None
df.iloc[3,3] = np.nan
df.iloc[5,2] = None
df.iloc[2,1] = None

df
```

```python
df.isnull()
```

```python
df.loc[[True,True,False,False, True, False, True, True]]
```

```python
df.notnull()
```

```python
df.notnull().any(axis=1) # 只要有true 则返回true
df.notnull().all(axis=1) # 所有为true 返回true
```

```python
df[df.notnull().all(axis=1)]
```

```python
# 固定搭配
# is_null =>  any
# not_null => all
# 结论:  df.notnull().all(axis=1)作为元数据的行索引，就可以将对应的空行删除
```

```python
indexs = df.loc[df.isnull().any(axis=1)].index # 获取的是值对应的行索引
indexs
df.drop(labels=indexs, axis=0) # axis 0 表示行， 1表示列
```

```python
df.dropna(axis=0) # 直接过滤行还是列， axis 0 表示行， 1表示列
```

```python
df.fillna(method='bfill', axis=0)
```
