```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
df1 = DataFrame(data=np.random.randint(0,100,size=(3,3)), index=['A','B','C'], columns=['a','b','c'])
df2 = DataFrame(data=np.random.randint(0,100,size=(3,3)), index=['A','B','D'], columns=['a','b','d'])

pd.concat((df1, df1), axis=0)
```

```python
pd.concat((df1,df2), axis=1, sort=False)
```

```python
pd.concat((df1,df2), axis=1, join='inner')
```

```python
df1 = DataFrame({
    'employee' : ['Bob', 'Jake', 'Lisa'],
    'group' : ['Accouting','Engineering','Engineering']
})
df1
```

```python
df2 = DataFrame({
    'employee' : ['Lisa', 'Bob', 'Jake'],
    'hire_date' : [2004,2008,2012]
})
df2
```

```python
pd.merge(df1,df2)
```

```python
df1 = DataFrame({
    'employee' : ['Bob', 'Jake', 'Lisa'],
    'group' : ['Accouting','Engineering','Engineering']
})
df1
```

```python
df5= DataFrame({
    'group' : ['Engineering','Engineering','HR'],
    'supervisor': ['Carly', 'Guido', 'Steve']
})
df5
```

```python
pd.merge(df1,df5,how='outer',on='group')
```

```python
pd.merge(df1,df5,how='inner',on='group')
```

```python
pd.merge(df1,df5,how='left',on='group')
```

```python
pd.merge(df1,df5,how='right',on='group')
```

```python
df1 = DataFrame({
    'employee' : ['Jack', 'Summer', 'Steve'],
    'group' : ['Accouting','Finance','Marketing']
})
df1
```

```python
df2 = DataFrame({
    'employee' : ['Jack', 'Bob', 'Jake'],
       'hire_date' : [2003,2009,2012],
    'group' : ['Accouting','sell','ceo']
})
df2
```

```python
pd.merge(df1, df2, on='group')
```

```python
df1 = DataFrame({
    'employee' : ['Bobs', 'Linda', 'Bill'],
    'group' : ['Accouting','Product','Marketing'],
       'hire_date' : [1998,2017,2018],
})
df1
```

```python
df5 = DataFrame({
    'name' : ['Lisa', 'Bobs', 'Bill'],
    'hire_date' : [1998,2016,2007],
})
df5
```

```python
pd.merge(df1,df5, left_on='employee', right_on='name')
```
