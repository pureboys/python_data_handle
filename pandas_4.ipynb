```python
dic =  {
    'name' : ['周杰伦', '张三', '周杰伦'],
    'salary' : [15000, 20000, 15000]
}

df = DataFrame(data=dic)
df
```

```python
# 映射关系表
dic = {
    '周杰伦' : 'jay',
    '张三' : 'tom'
}
df['e_name'] = df['name'].map(dic)
df
```

```python
def after_sal(s):
    return s-(s-3000) * 0.5
# 超过3000的部分钱进行缴纳50%的税
df['after_sal'] = df['salary'].map(after_sal)
df
```

```python
df = DataFrame({
    'item' : ['Apple', 'Banana', 'Orange', 'Banana', 'Orange', 'Apple'],
    'price' : [4,3,3,2.5,4,2],
    'color' : ['red', 'yellow', 'yellow', 'green', 'green', 'green'],
    'weight' : [12,20,50,30,20,44]
})
df
```

```python
df.groupby(by='item', axis=0)
```

```python
df.groupby(by='item', axis=0).groups
```

```python
# 给df 创建一个新列， 内容为哥哥水果的评价价格
df.groupby(by='item',axis=0).mean()['price']
mean_price_s = df.groupby(by='item',axis=0)['price'].mean()
```

```python
dic  = mean_price_s.to_dict()
df['item'].map(dic)
```

```python
df['mean_price'] = df['item'].map(dic)
df
```

```python
# 按照颜色的水果的平均价格
color_mean_price_s = df.groupby(by='color', axis=0)['price'].mean()
```

```python
color_mean_price_s
```

```python
dic = color_mean_price_s.to_dict()
dic
```

```python
df['mean_price_color'] = df['color'].map(dic)
df
```

```python
def my_mean(p):
    sum = 0
    for i in p :
        sum += i
    return sum/len(p)
    
```

```python
df.groupby(by='item')['price'].apply(my_mean)
```

```python
df.groupby(by='item')['price'].transform(my_mean)
```
